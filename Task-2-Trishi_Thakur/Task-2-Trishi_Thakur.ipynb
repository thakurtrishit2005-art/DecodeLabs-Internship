```python
# Core data handling and array manipulation libraries
import numpy as np
import pandas as pd

# Data visualization frameworks
import matplotlib.pyplot as plt
import seaborn as sns

# Built-in dataset and machine learning pipeline tools from Scikit-Learn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Environment ready! All data science toolkits loaded successfully.")
```

```python
# 1. Load the raw Iris benchmark data
iris = load_iris()

# 2. Convert to a Pandas DataFrame so it's readable as a clean table
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

# 3. Map numerical target classes (0, 1, 2) to their actual biological names
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species_name'] = df['target'].map(species_map)

# 4. Display dataset metrics
print(f"Dataset Summary: {df.shape[0]} balanced flower samples found.")
print(f"Target Classes to predict: {list(species_map.values())}\n")
print("First 5 entries of our tabular data view:")
df.head()
```

```python
# Extract features (X) and true labels (y)
X = iris.data
y = iris.target

# Split data: 80% used for model training logic, 20% reserved for strict evaluation
# We use random_state=10 to ensure our split shuffles nicely and exposes overlap
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=10, shuffle=True
)

print("Structural Integrity Check:")
print(f"   • Training Subset (Pattern Recognition): {X_train.shape[0]} rows")
print(f"   • Testing Subset (Validation Evaluation): {X_test.shape[0]} rows")
```

```python
# Initialize the scaler to force Mean = 0 and Variance = 1
scaler = StandardScaler()

# CRITICAL ML PRACTICE: Fit ONLY on training data to strictly prevent data leakage
X_train_scaled = scaler.fit_transform(X_train)

# Transform the test data using parameters learned from the training data
X_test_scaled = scaler.transform(X_test)

print("Gatekeeper Rule complete: All dimensions scaled via StandardScaler.")
```

```python
error_rates = []

# Test all K neighborhood values from 1 to 20
for k in range(1, 21):
    knn_test = KNeighborsClassifier(n_neighbors=k)
    knn_test.fit(X_train_scaled, y_train)
    predictions_k = knn_test.predict(X_test_scaled)
    error_rates.append(np.mean(predictions_k != y_test))

# Draw the Elbow Plot curve
plt.figure(figsize=(10, 5))
plt.plot(range(1, 21), error_rates, color='#1f77b4', linestyle='--', marker='o', markerfacecolor='red', markersize=8)
plt.title('Tuning the Engine: Error Rate vs. chosen K Value', fontsize=12, weight='bold')
plt.xlabel('Number of Neighbors (K Value)')
plt.ylabel('Error Rate')
plt.grid(True, alpha=0.5)
plt.show()

print("Analysis: The 'Elbow' chart reveals that K=5 minimizes error cleanly right before spikes occur.")
```

```python
# 1. Instantiate (Build the mathematical framework frame)
final_model = KNeighborsClassifier(n_neighbors=5)

# 2. Fit (Let the machine process history and derive decision boundaries)
final_model.fit(X_train_scaled, y_train)

# 3. Predict (Apply learned algorithmic logic to brand-new, unseen samples)
final_predictions = final_model.predict(X_test_scaled)

print("KNN Model training cycle fully locked and finalized.")
```

```python
# Compute metrics
accuracy = accuracy_score(y_test, final_predictions)
matrix = confusion_matrix(y_test, final_predictions)

# Plot a polished, publication-ready visual grid map
plt.figure(figsize=(7, 5))
sns.heatmap(
    matrix, 
    annot=True, 
    fmt='d', 
    cmap='Blues', 
    xticklabels=iris.target_names, 
    yticklabels=iris.target_names,
    square=True,
    annot_kws={"size": 13, "weight": "bold"},
    linewidths=1,
    linecolor='white'
)

plt.title(f'Diagnostic Tool: Confusion Matrix Grid\n(Overall Raw Accuracy: {accuracy * 100:.2f}%)', fontsize=12, weight='bold', pad=15)
plt.xlabel('AI Logic Predictions', fontsize=10, weight='bold', labelpad=10)
plt.ylabel('True Historical Class', fontsize=10, weight='bold', labelpad=10)
plt.show()

# Print the formal classification table block
print("\n" + "="*60)
print("       STRATEGIC TRADE-OFFS: DETAILED PERFORMANCE REPORT")
print("="*60)
print(classification_report(y_test, final_predictions, target_names=iris.target_names))
print("="*60)
```
